## Part 1

We'll want regexps for the input, and arrays for the sleep.

```python
import re

import numpy as np
```

```python
testInput_str='''
[1518-11-01 00:00] Guard #10 begins shift
[1518-11-01 00:05] falls asleep
[1518-11-01 00:25] wakes up
[1518-11-01 00:30] falls asleep
[1518-11-01 00:55] wakes up
[1518-11-01 23:58] Guard #99 begins shift
[1518-11-02 00:40] falls asleep
[1518-11-02 00:50] wakes up
[1518-11-03 00:05] Guard #10 begins shift
[1518-11-03 00:24] falls asleep
[1518-11-03 00:29] wakes up
[1518-11-04 00:02] Guard #99 begins shift
[1518-11-04 00:36] falls asleep
[1518-11-04 00:46] wakes up
[1518-11-05 00:03] Guard #99 begins shift
[1518-11-05 00:45] falls asleep
[1518-11-05 00:55] wakes up
'''
```

```python
input_ls=testInput_str.strip().split('\n')
input_ls.sort()
input_ls
```

```python
with open('inputs/day4') as fIn:
    input_ls=fIn.read().strip().split('\n')
input_ls.sort()
input_ls
```

Go through each case, and parse each line. Note that for the actual input we need to sort it, but AFAICT we should just be able to sort as strings. However, using the test input for the moment:

```python
guard_sleep_dict={}

for nl in input_ls:
    # Test for new guard
    m=re.match('\[(\d\d\d\d-\d\d-\d\d) (\d\d):(\d\d)\] Guard #(\d+) begins shift', nl)
    if m:
        (date, time_hh, time_mm, guard_id)=m.groups()
        if guard_id not in guard_sleep_dict:
            guard_sleep_dict[guard_id]={}
        if date not in guard_sleep_dict[guard_id]:
            guard_sleep_dict[guard_id][date]=np.zeros(60, dtype=np.int)
            
    # Assumes that a wakes up 
    m=re.match('\[(\d\d\d\d-\d\d-\d\d) (\d\d):(\d\d)\] falls asleep', nl)
    if m:
        sleep_i=int(m.group(3))
    m=re.match('\[(\d\d\d\d-\d\d-\d\d) (\d\d):(\d\d)\] wakes up', nl)
    if m:
        wake_i=int(m.group(3))
        guard_sleep_dict[guard_id][date][sleep_i:wake_i]=1

guard_sleep_dict
```

OK, now we've got that, we can find out which guard spends the longest asleep:

```python
total_sleep_dict={}
for g in guard_sleep_dict:
    total_sleep_dict[g]=sum([sum(ar) for ar in guard_sleep_dict[g].values()])
total_sleep_dict
```

```python
longest_sleeper_id=[x for x in total_sleep_dict if total_sleep_dict[x]==max(total_sleep_dict.values())][0]
longest_sleeper_id
```

And finally get the minute for which they are asleep the longest:

```python
sum(guard_sleep_dict[longest_sleeper_id].values())
```

```python
longest_sleeping_minute=np.argmax(sum(guard_sleep_dict[longest_sleeper_id].values()))
longest_sleeping_minute
```

```python
longest_sleeping_minute*int(longest_sleeper_id)
```

## Part 2

This one shouldn't be too bad. If we get the distribution for each guard:

```python
sleep_distributions={id:sum(guard_sleep_dict[id].values()) for id in guard_sleep_dict}
sleep_distributions
```

And convert to a list:

```python
[(max(sleep_distributions[i]), np.argmax(sleep_distributions[i]), i) for i in sleep_distributions]
```

And take the triple with the largest first value:

```python
(_, minute, guard_id)=sorted([(max(sleep_distributions[i]), 
                               np.argmax(sleep_distributions[i]), 
                               i) for i in sleep_distributions], reverse=True)[0]
```

```python
minute*int(guard_id)
```
