Want a function that takes a string representing the encrypted name, and returns a dataframe containing each of the letters and the number of times it appears.

```python
def name_to_df(stringIn):
    ks=Counter(stringIn)
    cs=[k for k in ks.keys() if k!='-']
    return pd.DataFrame({'char':cs, 'count':[ks[c] for c in cs]})

name_to_df('not-a-real-room-')
```

```python
def check_room(stringIn):
    m=re.match('(?P<code>\D+)(?P<id>\d+)\[(?P<checksum>.....)\]', stringIn)
    
    # First get the checksum for this code
    t_df=name_to_df(m.group('code'))
    
    t_df.sort_values(['count', 'char'], axis=0, ascending=[False, True], inplace=True)
    
    
    
    return {'id':int(m.group('id')),
            'real':m.group('checksum')==''.join(t_df['char'].iloc[0:5]),
            'code':m.group('code')}
```
